```python
import numpy as np
import matplotlib as mpt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
y1 = (1+np.cos(x*2)*np.exp(-x/10))
y2 = (1+np.cos(x*20)/20 + np.sin((x+np.pi/3)*30)/20)+np.random.random(len(x))/20
f = y1+y2
```

```python
plt.plot(x,f)
```

```python
plt.plot(x,f)
plt.xlim(-2,-1.0)
plt.ylim(0.7,1.3)
```

```python
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

with sns.plotting_context('talk', font_scale=1.5):
    sns.set_style('white')
    fig, axes = plt.subplots(1)
    
    axes.plot(x, f)
    plt.xticks(visible=False)
    plt.yticks(visible=False)
#     plt.setp(axins,xticks=[],yticks=[])
    plt.ylabel('Free energy')
    plt.xlabel('Conformational coordinate')
    
    axes.text(0.15, 0.95, 'A',
        verticalalignment='top', horizontalalignment='left',
        transform=axes.transAxes, fontsize=30)
    
    
    axins = zoomed_inset_axes(axes,2,loc=1) # zoom=6
    axins.plot(x,f, zorder=0)
    x1,x2,y1,y2 = -2,-1,0.7,1.3
    axins.set_xlim(x1,x2)
    axins.set_ylim(y1,y2)
    plt.xticks(visible=False)
    plt.yticks(visible=False)
    plt.setp(axins,xticks=[],yticks=[])
    
    mark_inset(axes, axins, loc1=2, loc2=4, fc="none", ec="k" , color='k', lw=1, zorder=1)
    
    axins.text(0.15, 0.95, 'B',
        verticalalignment='top', horizontalalignment='left',
        transform=axins.transAxes, fontsize=30)
    
    plt.savefig('figures/hierarchy.pdf')

#     axes[0].plot(x,y1+y2)
#     axes[1].plot(x,y1+y2)
#     axes[1].set_xlim(-2,-1)
#     axes[1].set_ylim(0.7,1.3)
```
