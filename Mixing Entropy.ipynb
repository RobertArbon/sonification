```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import copy
```

```python
hmm_traj = np.load('data/hmm_trajectories/traj-0.npy')
```

```python
hmm_sorted = copy.deepcopy(hmm_traj)
hmm_sorted.sort(axis=1)
hmm_sorted = hmm_sorted[:, 1:]

```

```python
hmm_sorted
```

```python
entropy = np.sum(hmm_sorted*np.log(hmm_sorted), axis=1)
```

```python
plt.plot(hmm_traj[100:250:1, 0], c='g')
plt.plot(hmm_traj[100:250:1, 1], c='y')
plt.plot(hmm_traj[100:250:1, 2], c='k')
```

```python
scaled_entropy = entropy-np.min(entropy)
```

```python
plt.plot(scaled_entropy)

```

```python
plt.plot(scaled_entropy[100:250])
```
