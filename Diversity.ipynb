```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import copy
```

```python
fname = 'Data/Chodera_data/Processed/probabilistic_traj_lag1.0ps.npy'
```

```python
hmm_traj = np.load(fname)
```

```python
hmm_sorted = copy.deepcopy(hmm_traj)
hmm_sorted.sort(axis=1)
hmm_sorted = hmm_sorted[:, -2:]
```

```python
hmm_sorted
```

```python
entropy = np.sum(hmm_sorted*np.log(hmm_sorted), axis=1)
```

```python
plt.plot(hmm_traj[100:250:1, 0], c='g')
plt.plot(hmm_traj[100:250:1, 1], c='y')
plt.plot(hmm_traj[100:250:1, 2], c='k')
plt.plot(hmm_traj[100:250:1, 3], c='k')
```

```python
scaled_entropy = entropy-np.min(entropy)
```

```python
plt.plot(scaled_entropy)
```

```python
plt.plot(scaled_entropy[100:250])
```

```python
np.save(fname.replace('probabilistic', 'mixing_entropy'), scaled_entropy)
```
